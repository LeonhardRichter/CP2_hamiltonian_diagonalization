```python
import pickle
import matplotlib.pyplot as plt

import numpy as np
```

```python
with open("parameters_optimization_data.pickle", "rb") as file:
    data = pickle.load(file)
```

```python
def plot_test(test: str):
    series = data[test]
    ref_run = series[(30, 100, 0.0001)]
    for k, run in series.items():
        series[k]["distance"] = np.linalg.norm(
            run["final vector"][0] - ref_run["final vector"][0]
        )

    print([series[k]["distance"] for k in series.keys()])

    parameter_index = {
        "test cut_off": 0,
        "test iterations": 1,
        "test time_step": 2,
    }

    x = list()
    y = list()
    z = list()
    for k, v in series.items():
        x.append(k[parameter_index[test]])
        y.append(v["distance"])
        z.append(v["duration"])

    x = np.array(x)
    y = np.array(y)
    z = np.array(z)

    plt.scatter(x, y, s=2)
    plt.show()
    plt.scatter(x, z, s=2)
    plt.show()
    plt.scatter(x, x * z, s=2)
    plt.show()


plot_test("test iterations")
```
